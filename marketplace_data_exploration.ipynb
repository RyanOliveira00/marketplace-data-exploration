## Decisões

### 1 - Melhoria na qualidade de servico
Implementar um processo automatizado para identificar e descontinuar produtos que, apesar de terem um alto volume de pedidos (mais de 40), consistentemente recebem avaliações baixas (1 ou 2) dos clientes.



```python
orders_com_reviews = pd.merge(df_order_items, df_reviews, on='order_id', how='inner')
orders_com_reviews_com_products = pd.merge(orders_com_reviews, df_products, on='product_id', how='inner')

reviews_com_1_ou_2 = orders_com_reviews_com_products[orders_com_reviews_com_products['review_score'] <= 2].fillna('')

produtos_com_reviews_1_ou_2 = reviews_com_1_ou_2.groupby('product_id').agg({'review_score': 'mean', 'order_id': 'count', 'review_comment_message': ' , '.join}).sort_values(by='order_id', ascending=False).reset_index()

produtos_com_reviews_1_ou_2
```

```python
top_10_produtos_com_reviews_1_ou_2 = produtos_com_reviews_1_ou_2.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_produtos_com_reviews_1_ou_2['product_id'], top_10_produtos_com_reviews_1_ou_2['order_id'], color='red')
plt.xlabel('Número de Avaliações Baixas')
plt.ylabel('ID do Produto')
plt.title('Top 10 Produtos com Maior Número de Avaliações Baixas')
plt.gca().invert_yaxis()
plt.show()
```

```python
comments_by_score = df_reviews.groupby('review_score')['review_comment_message'].count().sort_values(ascending=False)
comments_by_score.plot(kind='bar')
plt.title('Número de Comentários por Pontuação de Revisão')
plt.xlabel('Pontuação de Revisão')
plt.ylabel('Número de Comentários')
plt.show()
```

Métricas para acompanhar:

1. **Número de Produtos Descontinuados**: Esta métrica rastreará quantos produtos foram identificados e removidos do inventário com base nos critérios estabelecidos.

2. **Avaliação Média dos Produtos**: Acompanhar a avaliação média dos produtos ao longo do tempo pode ajudar a entender se a qualidade geral dos produtos está melhorando após a implementação desta decisão.

3. **Número de Pedidos de Produtos com Avaliações Baixas**: Esta métrica rastreará o número de pedidos de produtos que têm avaliações baixas. Uma diminuição nesta métrica ao longo do tempo pode indicar que a decisão está tendo um impacto positivo.

4. **Satisfação do Cliente**: Acompanhar a satisfação geral do cliente ao longo do tempo pode ajudar a entender se a remoção de produtos com avaliações baixas está tendo um impacto positivo na experiência do cliente.

### 2 - Otimizar a logística de entrega


Expandir a base de vendedores ao redor do pais ou construir cds para melhor atender a demanda dos clientes e melhorar a distribuição geográfica dos vendedores.



```python
clientes_geo = pd.merge(df_customers, df_geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
vendedores_geo = pd.merge(df_sellers, df_geolocation, left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')

amostra_clientes = clientes_geo.sample(n=9999999)
amostra_vendedores = vendedores_geo

plt.figure(figsize=(12, 6))
sns.scatterplot(x='geolocation_lng', y='geolocation_lat', data=amostra_clientes, label='Clientes')
sns.scatterplot(x='geolocation_lng', y='geolocation_lat', data=amostra_vendedores, label='Vendedores')
plt.title('Distribuição Geográfica de Clientes e Vendedores')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
customers_by_city = df_customers['customer_city'].value_counts().sort_values(ascending=False)
customers_by_city[:20].plot(kind='bar')  # Top 20 cidades
plt.title('Número de Clientes por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Número de Clientes')
plt.show()
```

```python
top_15_cidades_vendedores = vendedores_geo['seller_city'].value_counts().head(15)

plt.figure(figsize=(15, 6))
plt.bar(top_15_cidades_vendedores.index, top_15_cidades_vendedores.values, color='green')
plt.xticks(rotation=90)
plt.xlabel('Cidade')
plt.ylabel('Número de Vendedores')
plt.title('Top 15 Cidades com Mais Vendedores')
plt.show()
```

```python
df_merged = pd.merge(df_orders, df_customers, on='customer_id')
df_merged['delivery_time'] = pd.to_datetime(df_merged['order_delivered_customer_date']) - pd.to_datetime(df_merged['order_purchase_timestamp'])
avg_delivery_time = df_merged.groupby('customer_state')['delivery_time'].mean().sort_values()
avg_delivery_time.plot(kind='bar')
plt.title('Tempo Médio de Entrega por Estado')
plt.xlabel('Estado')
plt.ylabel('Tempo Médio de Entrega')
plt.show()
```

3. **Tempo de Entrega**: Com mais vendedores ou cds, o tempo de entrega para pedidos nesta região deve diminuir. Acompanhe esta métrica para confirmar se isso está ocorrendo.

2 - Receita total por categoria: Além do número de vendas, você pode querer acompanhar a receita gerada por cada categoria de produto. Isso pode ser calculado multiplicando o número de vendas pelo preço de cada produto.

3 - Número de devoluções por categoria: Para ter uma visão completa do desempenho de cada categoria, você pode querer acompanhar o número de devoluções. Isso pode indicar problemas com certos produtos ou categorias.

4 - Número de revisões e pontuação média por categoria: As revisões dos clientes podem fornecer insights valiosos sobre a satisfação do cliente com diferentes categorias de produtos.

5 - Vendas ao longo do tempo por categoria: Acompanhar as vendas ao longo do tempo pode ajudar a identificar tendências ou padrões sazonais em diferentes categorias de produtos.

## 5 - **Planejar estoque e logística para promoções**



```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])

promo_periods = [
    {'start': '2017-11-24', 'end': '2017-11-27'},
]

sales_during_promos = []
for period in promo_periods:
    start_date = pd.to_datetime(period['start'])
    end_date = pd.to_datetime(period['end'])
    sales = df_orders[(df_orders['order_purchase_timestamp'] >= start_date) & (df_orders['order_purchase_timestamp'] <= end_date)].shape[0]
    sales_during_promos.append({'start': start_date, 'end': end_date, 'sales': sales})

df_sales_during_promos = pd.DataFrame(sales_during_promos)
df_sales_during_promos.plot(x='start', y='sales', kind='bar')
plt.title('Número de Vendas Durante Períodos Promocionais')
plt.xlabel('Data de Início do Período Promocional')
plt.ylabel('Número de Vendas')
plt.show()
```

1 - Vendas por Cliente Durante Períodos Promocionais
2 - Margem de Lucro Durante Períodos Promocionais
3 - Vendas Totais Durante Períodos Promocionais
4 - Satisfação de usar o nosso sistemas em Períodos Promocionais
